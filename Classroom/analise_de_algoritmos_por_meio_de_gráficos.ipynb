# Global functions

```python
def retorna_tempo_e_contador(algoritmo, arranjo):
    tempo_inicial = time.time()
    (arranjo, contador) = algoritmo(arranjo)
    tempo_final = time.time()
    tempo_total = tempo_final - tempo_inicial
    return tempo_total, contador
    
def gera_arranjos_aleatorios(n_elementos):
    return np.random.randint(1, 999, n_elementos)


def gera_arranjos_decrescente(n_elementos):
    dec = np.arange(n_elementos)
    dec = dec[::-1]  # inverte a ordem
    return dec


def gera_arranjos_crescente(n_elementos):
    cresc = np.arange(n_elementos)
    return cresc


# insertion_sort(gera_arranjos_decrescente(10))
# insertion_sort(gera_arranjos_crescente(10))
# insertion_sort(gera_arranjos_aleatorios(10))
def calcula_tempo_algoritmo(algoritmo):
    aleatorio_tempo = []
    aleatorio_iteracoes = []
    crescente_tempo = []
    crescente_iteracoes = []
    decrescente_tempo = []
    decrescente_iteracoes = []

    for i in range(1,1001,5):
        (tempo,contador) = retorna_tempo_e_contador(algoritmo,gera_arranjos_aleatorios(i))
        aleatorio_tempo.append(tempo)
        aleatorio_iteracoes.append(contador)

        (tempo,contador) = retorna_tempo_e_contador(algoritmo,gera_arranjos_crescente(i))
        crescente_tempo.append(tempo)
        crescente_iteracoes.append(contador)

        (tempo,contador) = retorna_tempo_e_contador(algoritmo,gera_arranjos_decrescente(i))
        decrescente_tempo.append(tempo)
        decrescente_iteracoes.append(contador)

    aleatorio = {'tempo': aleatorio_tempo, 'iteracoes': aleatorio_iteracoes}
    crescente = {'tempo': crescente_tempo, 'iteracoes': crescente_iteracoes}
    decrescente = {'tempo': decrescente_tempo, 'iteracoes': decrescente_iteracoes}

    return {'aleatorio': aleatorio, 'crescente': crescente, 'decrescente': decrescente}

def gera_graficos(algoritmo):
    data = calcula_tempo_algoritmo(algoritmo)
    aleatorio = data['aleatorio']
    crescente = data['crescente']
    decrescente = data['decrescente']

    plt.figure(figsize=(10,8))
    plt.plot(aleatorio['tempo'],'k',label='aleatorio')
    plt.plot(crescente['tempo'],'g',label='crescente')
    plt.plot(decrescente['tempo'],'r',label='decrescente')
    plt.xlabel('Numero de entradas')
    plt.ylabel('Tempo')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,8))
    plt.plot(aleatorio['iteracoes'],'k',label='aleatorio')
    plt.plot(crescente['iteracoes'],'g',label='crescente')
    plt.plot(decrescente['iteracoes'],'r',label='decrescente')
    plt.xlabel('Numero de entradas')
    plt.ylabel('Interações')
    plt.legend()
    plt.show()

```

# Insertion Sort

```python
def insertion_sort(arranjo):
    contador = 0
    for j in range(1, len(arranjo)):
        chave = arranjo[j]
        i = j - 1
        contador += 1
        while i >= 0 and arranjo[i] > chave:
            arranjo[i + 1] = arranjo[i]
            i -= 1
            contador += 1
        contador += 1
        arranjo[i + 1] = chave
    contador += 1
    return arranjo, contador

gera_graficos(insertion_sort)
```

# Selection Sort

```python
def selectionSort(array):
    contador = 0
    for i in range(0,len(array)):
        menor = i
        contador+=1
        for j in range(i+1,len(array)):
            contador += 1
            if array[menor] > array[j]:
                menor = j
                contador += 1
        contador += 1
        (array[i],array[menor]) = (array[menor], array[i])
    contador+=1
    return array, contador

gera_graficos(selectionSort)
```

# Bubble Sort WITH FLAG

$ \Omega (n)$ e $ O (n^2) $


```python
def bubbleSort(array):
    fim = len(array) - 1    
    contador = 0     
    alterado = False      
    for i in range(0,fim): 
        contador += 1       
        if (alterado == True):
            alterado = False
        for j in range(0, fim-i):
            contador += 1
            if array[j] > array[j + 1]:
                (array[j],array[j+1]) = (array[j+1], array[j])
                alterado = True
        contador += 1
        if not alterado: 
            return array, contador
        
        contador += 1
        fim -= 1
    return array, contador
    
gera_graficos(bubbleSort)
```

# Bubble Sort NO FLAG

```python
def bubbleSortNotFlag(array):
    fim = len(array) - 1    
    contador = 0     
    for i in range(0,fim): 
        contador += 1       
        for j in range(0, fim):
            contador += 1
            if array[j] > array[j + 1]:
                contador += 1
                (array[j],array[j+1]) = (array[j+1], array[j])
        contador += 1
    contador += 1
    return array, contador
    
gera_graficos(bubbleSortNotFlag)
```
